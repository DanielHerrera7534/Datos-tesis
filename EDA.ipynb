```python
df_geo = df[
    (df["Latitude"].between(41.6, 42.1)) &
    (df["Longitude"].between(-87.95, -87.5))
]
```

```python
plt.figure(figsize=(6,6))

plt.scatter(
    df_geo["Longitude"],
    df_geo["Latitude"],
    s=1,
    alpha=0.3
)

plt.title("Ubicación de crímenes en Chicago")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.show()
```

```python
import matplotlib.pyplot as plt

def crime_hotspot(crime_type):

    data = df_geo[df_geo["Primary Type"] == crime_type]

    plt.figure(figsize=(8,8))

    plt.hexbin(
        data["Longitude"],
        data["Latitude"],
        gridsize=120,
        cmap="Reds",
        mincnt=1,
        bins="log"
    )

    plt.colorbar(label="Número de crímenes")
    plt.title(f"Hotspots de {crime_type} en Chicago")

    plt.xlabel("Longitude")
    plt.ylabel("Latitude")

    plt.gca().set_facecolor("lightgray")

    plt.show()
```

```python
top_crimes = df_geo["Primary Type"].value_counts().head(4).index

for crime in top_crimes:
    crime_hotspot(crime)
```

```python
top_crimes = df_geo["Primary Type"].value_counts().head(4).index

fig, axes = plt.subplots(2,2, figsize=(12,12))

for ax, crime in zip(axes.flatten(), top_crimes):

    data = df_geo[df_geo["Primary Type"] == crime]

    hb = ax.hexbin(
        data["Longitude"],
        data["Latitude"],
        gridsize=100,
        cmap="Reds",
        bins="log",
        mincnt=1
    )

    ax.set_title(crime)
    ax.set_facecolor("lightgray")

plt.tight_layout()
plt.show()
```
